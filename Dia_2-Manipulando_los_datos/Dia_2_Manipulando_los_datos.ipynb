```python
prestamos_completo.head()
```

## Asignar la columna con la nomenclatura de la  Clasificación Decimal Universal (CDU)

```python
CDU_lista = []
for CDU in prestamos_completo['localizacion']:
  if(CDU < 100):
    CDU_lista.append('Generalidades')
  elif(CDU < 200):
    CDU_lista.append('Filosofía y psicología')
  elif(CDU < 300):
    CDU_lista.append('Religión')
  elif(CDU < 400):
    CDU_lista.append('Ciencias sociales')
  elif(CDU < 500):
    CDU_lista.append('Clase vacía')
  elif(CDU < 600):
    CDU_lista.append('Matemáticas y ciencias naturales')
  elif(CDU < 700):
    CDU_lista.append('Ciencias aplicadas')
  elif(CDU < 800):
    CDU_lista.append('Bellas artes')
  elif(CDU < 900):
    CDU_lista.append('Lenguaje')
  else:
    CDU_lista.append('Geografía. Biografía. Historia.')

prestamos_completo['CDU_general'] = CDU_lista
```

```python
prestamos_completo.head()
```

## Excluir la columna "registro_sistema"

```python
prestamos_completo.drop(columns=['registro_sistema'],inplace=True)
```

## Modificar tipo de dato de la matrícula

```python
prestamos_completo['matricula_o_siap'] = prestamos_completo['matricula_o_siap'].astype('string')
```
