## Asignar la columna con la nomenclatura de la  Clasificación Decimal Universal (CDU)

```python
CDU_lista = []
for CDU in prestamos_completo['localizacion']:
  if(CDU < 100):
    CDU_lista.append('Generalidades')
  elif(CDU < 200):
    CDU_lista.append('Filosofía y psicología')
  elif(CDU < 300):
    CDU_lista.append('Religión')
  elif(CDU < 400):
    CDU_lista.append('Ciencias sociales')
  elif(CDU < 500):
    CDU_lista.append('Clase vacía')
  elif(CDU < 600):
    CDU_lista.append('Matemáticas y ciencias naturales')
  elif(CDU < 700):
    CDU_lista.append('Ciencias aplicadas')
  elif(CDU < 800):
    CDU_lista.append('Bellas artes')
  elif(CDU < 900):
    CDU_lista.append('Lenguaje')
  else:
    CDU_lista.append('Geografía. Biografía. Historia.')

prestamos_completo['CDU_general'] = CDU_lista
```

## Cantidad de ejemplares prestados por año

```python
prestamos_fecha = pd.DataFrame(prestamos_completo['fecha_prestamo'].value_counts()).reset_index()
prestamos_fecha.columns = ['fecha','cantidad']
prestamos_fecha['fecha'] = pd.to_datetime(prestamos_fecha['fecha'],format='mixed')
prestamos_fecha
```

```python
prestamos_anual = prestamos_fecha.groupby(by=prestamos_fecha.fecha.dt.year).count()
prestamos_anual.index.name = 'año'
prestamos_anual = prestamos_anual[['cantidad']]
prestamos_anual
```

## Configurando el tema de los gráficos

```python
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='deep',
              font_scale=1.3,
              rc={"figure.figsize":(15,8)})
```

```python
ax = sns.lineplot(data=prestamos_anual,x=prestamos_anual.index,y='cantidad')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x', rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title('Cantidad de ejemplares prestados por año'+'\n',size=20,loc='left',weight='bold')

ax=ax
```

---
Se puede percibir que desde 2018 el número de ejemplares prestados por las bibliotecas ha ido disminuyendo, con una caída muy acentuada en 2024.

El año 2025 no puede evaluarse de la misma manera, debido al período de la cantidad de datos.

## ¿Cuántos préstamos hubo por cada mes?

```python
prestamos_por_mes = prestamos_fecha.groupby(by=prestamos_fecha.fecha.dt.month).count()
prestamos_por_mes.index.name = 'mes'
prestamos_por_mes = prestamos_por_mes[['cantidad']]
prestamos_por_mes
```

## Modificando el formato de los meses

```python
diccionario_meses = {1:'Ene',2:'Feb',3:'Mar',4:'Abr',
                    5:'May',6:'Jun',7:'Jul',8:'Ago',
                    9:'Sep',10:'Oct',11:'Nov',12:'Dic'}
```

```python
prestamos_por_mes.index = prestamos_por_mes.index.map(diccionario_meses)
```

```python
prestamos_por_mes
```

```python
ax = sns.lineplot(data=prestamos_por_mes,x='mes',y='cantidad')
ax.set(xlabel=None,ylabel=None)
ax.tick_params(axis='x',rotation=30)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))

ax.set_title("Cantidad de ejemplares prestados por mes"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2015 a 2025',x=-0.5,y=265000,fontsize=18, ha='left',color='gray')

ax=ax
```

Se puede observar que los meses con mayores números de ejemplares prestados fueron **marzo** y **agosto**.

Al tratarse de bibliotecas universitarias, los meses de vacaciones como enero, julio y diciembre presentan los números más bajos.

Lo cual no es ninguna novedad; sin embargo, si todo el personal se tomara vacaciones durante este período, no se podría mantener el funcionamiento completo de las bibliotecas.

Principalmente por actividades de inventario que pueden realizarse anualmente o cada dos años, y cuyo mejor período sería de diciembre a enero.

Por lo tanto, meses como mayo, junio, octubre y noviembre pueden ser una opción a considerar por las directivas de las bibliotecas.

También se nota una gran caída en el mes de junio. Cabe aquí una alerta para investigar más qué está ocurriendo y desarrollar estrategias y acciones de marketing de las bibliotecas para incrementar el número de ejemplares prestados.

Como por ejemplo: una acción que promueva préstamos de libros para lectura durante las vacaciones de julio.


## ¿Cuál es la cantidad de préstamos a lo largo del día?


```python
prestamos_fecha.fecha
```

```python
prestamos_por_hora = prestamos_fecha.groupby(by=prestamos_fecha.fecha.dt.hour).count()
prestamos_por_hora.index.name = 'hora'
prestamos_por_hora = prestamos_por_hora.reset_index()
prestamos_por_hora = prestamos_por_hora[['hora','cantidad']]
prestamos_por_hora
```

```python
prestamos_por_hora = prestamos_por_hora.sort_values(ascending=True,by='cantidad')
```

```python
ax = sns.barplot(data=prestamos_por_hora,y='cantidad',x='hora',
                 palette='Blues',hue='cantidad',dodge=False)  # Ordenar por la cantidad de ejemplares prestados
plt.legend([],[], frameon=False)                              # Excluir la leyenda del gráfico

ax.set(xlabel='Horario',ylabel=None)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',','.')))
ax.set_title("Cantidad de ejemplares prestados por franja horaria"+"\n",size=20,loc='left',weight='bold')
ax.text(s='Período de 2015 a 2025',x=-0.5,y=225000,fontsize=18, ha='left',color='gray')
ax=ax
```

---
A través del gráfico, se puede identificar que los períodos de las **10 h y 16 h son los de mayor pico en la atención de préstamos.**

Los períodos de **comienzo de la mañana** (entre 7h y 9h) y de la **noche** (entre 19h y 22h) pueden ser los mejores horarios para realizar otras actividades además de la atención al público.
